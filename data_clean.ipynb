```python
# read data
sales = pd.read_csv("Online_Sales.csv")
customers = pd.read_excel("CustomersData.xlsx")
tax = pd.read_excel("Tax_amount.xlsx")
discount = pd.read_csv("Discount_Coupon.csv")
marketing = pd.read_csv("Marketing_Spend.csv")
```

```python

# Tạo cột Month để join với Discount_Coupon
sales["Transaction_Date"] = pd.to_datetime(sales["Transaction_Date"])
sales["Month"] = sales["Transaction_Date"].dt.strftime("%b")  # "Jan", "Feb"...

# Join từng bảng vào sales
df = sales.merge(customers, on="CustomerID",         how="left")
df = df.merge(tax,          on="Product_Category",   how="left")
df = df.merge(discount,     on=["Product_Category", "Month"], how="left")

marketing["Transaction_Date"] = pd.to_datetime(marketing["Date"])
df = df.merge(marketing[["Transaction_Date", "Offline_Spend", "Online_Spend"]],
              on="Transaction_Date", how="left")

print("Shape:", df.shape)
df.columns.tolist()
```
